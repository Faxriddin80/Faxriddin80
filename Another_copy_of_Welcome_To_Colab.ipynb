```
import pandas as pd

df = pd.read_csv("toyota (3).csv")

df.columns = ["model","year","price","transmission","mileage",
              "fuel_type","tax","mpg","engine_size"]


print("Original rows:", len(df))
print("Missing values per column:\n", df.isna().sum())


df_clean = df.dropna().drop_duplicates().reset_index(drop=True)
print("After cleaning, rows:", len(df_clean))

df_clean.to_csv("toyota_clean.csv", index=False)
print("Cleaned dataset saved as toyota_clean.csv")
```

```
import pandas as pd


df = pd.read_csv("toyota (4).csv")


df.columns = ["model","year","price","transmission","mileage",
              "fuel_type","tax","mpg","engine_size"]


print("Original rows:", len(df))
print("Missing values per column:\n", df.isna().sum())


df_clean = df.dropna().drop_duplicates().reset_index(drop=True)
print("After cleaning, rows:", len(df_clean))


df_clean.to_csv("toyota_clean.csv", index=False)
print("Cleaned dataset saved as toyota_clean.csv")
```

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("toyota_clean.csv")


sns.set_style("whitegrid")
plt.figure(figsize=(10,5))
sns.histplot(df['price'], bins=50, kde=True, color='skyblue')
plt.title("Toyota Narx Taqsimoti")
plt.xlabel("Narx")
plt.ylabel("Son")
plt.show()


plt.figure(figsize=(10,5))
sns.barplot(x='year', y='price', data=df, ci=None, palette='viridis')
plt.title("Yil bo'yicha o'rtacha Narx")
plt.xlabel("Yil")
plt.ylabel("O'rtacha Narx")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(x='transmission', y='price', data=df, palette='Set2')
plt.title("Transmissiya bo'yicha Narx Taqsimoti")
plt.xlabel("Transmissiya")
plt.ylabel("Narx")
plt.show()


plt.figure(figsize=(10,6))
sns.scatterplot(x='mileage', y='price', data=df, hue='fuel_type', palette='Set1')
plt.title("Mileage vs Price (Fuel type bo'yicha)")
plt.xlabel("Mileage")
plt.ylabel("Price")
plt.show()
```
